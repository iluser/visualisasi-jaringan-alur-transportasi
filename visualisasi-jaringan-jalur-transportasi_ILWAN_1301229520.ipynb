```python
import pandas as pd
import networkx as nx
import numpy as np
import matplotlib.colors as mcolors
import matplotlib.cm as cm
import matplotlib.pyplot as plt
```

```python
df = pd.DataFrame({
    'from': [
        'Kebon Kelapa', 'Kebon Kelapa', 'Kebon Kelapa', 'Kebon Kelapa', 'Cibaduyut', 'Cicaheum',  'Cicaheum', 'Cicaheum', 'Ciroyom', 'Ciroyom', 'Ciroyom', 'Ciroyom', 'Dago', 'Dago', 'Margahayu Raya', 'Panghegar Permai', 'Sadang Serang', 'Sadang Serang', 'St Hall', 'St Hall', 'Stasiun Hall', 'Stasiun Hall', 'Stasiun Hall', 'Stasiun Hall', 'Stasiun Hall'

    ],
    'to': [
        'Cicaheum', 'Cicaheum (aceh)', 'Dago', 'Ledeng', 'Kebon Kelapa', 'Ciroyom', 'Ledeng', 'Leuwipanjang', 'Bumi Asri', 'Sarijadi', 'Sarijadi ', 'Bumi Asri', 'Pasar lnduk Caringin', 'Riung Bandung', 'Ledeng', 'Dipatiukur', 'Caringin', 'Caringin', 'Dago', 'Sadang Serang', 'Cirrrnbeuleurt', 'Cihampelas', 'Gede Bage', 'Gunung Batu', 'Sariladi'
    ]
})
G = nx.from_pandas_edgelist(df, 'from', 'to')
```

```python
def graph_draw_coloring(data=None, sizes='Black', colormap=None):
    if data != None:
        cent = np.fromiter(data.values(), float)
        sizes = cent / np.max(cent) * 200
        
        normalize = mcolors.Normalize(vmin=cent.min(), vmax=cent.max())
        colormap = cm.Set1_r

        scalarmappaple = cm.ScalarMappable(norm=normalize, cmap=colormap)
        scalarmappaple.set_array(cent)
        
    plt.figure(figsize=(20, 12))
    nx.draw_kamada_kawai(G, with_labels=True, node_size=12000, font_color="white", font_weight='300',  node_color=sizes, cmap=colormap)
    plt.show()
```

```python
graph_draw_coloring()
```

```python
beetwen = nx.betweenness_centrality(G)

df = pd.DataFrame(sorted(beetwen.items(), key=lambda p: p[1], reverse=True), columns=['Kota', 'Nilai Betweenness'])
df
```

```python
graph_draw_coloring(beetwen)
```

```python
degree = nx.degree_centrality(G)

df = pd.DataFrame(sorted(degree.items(), key=lambda p: p[1], reverse=True), columns=['Kota', 'Nilai Degree'])
df
```

```python
graph_draw_coloring(degree)
```
